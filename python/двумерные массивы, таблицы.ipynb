```python
data = [
["Петя", 15],
["Алексанр", 25],
["Аня", 3000],
]
print (f"|{'':>1}{'№ п/п':<6}", end="")
print (f"|{'':7}{'Имя':<10}", end="")
print (f"|{'':2}{'Оклад':<7}|")
a = 0
for i in data:
    a += 1
    print (f"|{a:>6} ", end="")
    print (f"|{i[0]:<17}", end="")
    print (f"|{i[1]:<9}|")    
```

```python
data=[
["Петя", 15],
["Саша", 25],
["Аня", 30],
]
print("="*36)
print(f"|{'':>2}{'№ п/п':<7}", end='')# сделали 2 оступа справ и слева, чтобы посчитать второй оттсут=отсутп+кол-во символов
print(f"|{'':>3}{'Имя':<7}", end='')
print(f"|{'':>4}{'Оклад':<9}|",)
a=0
print("="*36)
for i in data:
    a+=1
    print(f"|{a:>9}", end="") #9 отступов слева
    print(f"|{i[0]:<10}", end="") #10 отсупов справа
    print(f"|{i[1]:<13}|")
    print("_"*36)
```
