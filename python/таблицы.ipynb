```python
list = [
        ["000001", "Иванов ", 19999.85, 0.05],
        ["000150", "Никитин", 24999.56, 0.18],
        ["000160", "Жуков", 2500, 0.35],
        ["000030", "Васильев", 27999.9, 0.12],
        ["000045", "Сидоров", 25700.35, 0.3],
        ["000206", "Петров", 15000, 0.1]
]
print("="*60)
print(f"|{'':>2}{'Табельный номер':<17}", end='')# сделали 2 оступа справ и слева, чтобы посчитать второй оттсут=отсутп+кол-во символов
print(f"|{'':>5}{'ФИО':<8}", end='')
print(f"|{'':>3}{'Оклад':<8}", end='')
print(f"|{'':>2}{'Надбавка':<10}|",)
print("="*60)
for i in list:
    print(f"|{'':>6}{i[0]:<13}", end="")
    print(f"|{i[1]:<13}", end="")
    print(f"|{i[2]:>11}", end="")
    print(f"|{i[3]:12.0%}|")
    print("_"*60)
```

```python
list = [
        ["000001", "Иванов ", 19999.85, 0.05],
        ["000150", "Никитин", 24999.56, 0.18],
        ["000160", "Жуков", 2500, 0.35],
        ["000030", "Васильев", 27999.9, 0.12],
        ["000045", "Сидоров", 25700.35, 0.3],
        ["000206", "Петров", 15000, 0.1]
]
print("="*60)
print(f"|{'Табельный номер':^19}", end='')# сделали 2 оступа справ и слева, чтобы посчитать второй оттсут=отсутп+кол-во символов
print(f"|{'ФИО':^13}", end='')
print(f"|{'Оклад':^11}", end='')
print(f"|{'Надбавка':^12}|",)
print("="*60)
for i in list:
    print(f"|{i[0]:^19}", end="")
    print(f"|{i[1]:<13}", end="")
    print(f"|{i[2]:>11}", end="")
    print(f"|{i[3]:12.0%}|")
    print("_"*60)
```
