```python
age = int( input("Введите возраст"))
price = 1200
if age < 10 :
    print("Вам", age ,"лет и стоимость билета - ", price - price * 1, "руб.")
elif age < 18 :
    print("Вам", age ,"лет и стоимость билета - ", price - price * 0.2, "руб.")
elif age < 65 :
    print("Вам", age ,"лет и стоимость билета - ", price - price * 0, "руб.")
else :
    print("Вам", age ,"лет и стоимость билета - ", price - price * 0.5, "руб.")
```
